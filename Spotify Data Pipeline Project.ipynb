```python
# Importing Libraries

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
```

```python
# Initialize Spotipy with credentials
client_credentials_manager = SpotifyClientCredentials(
    client_id="bab7e460169e4297852bc7dc2614db8d",
    client_secret="ea1ef1b45d80443eb5dde8ae4662f7a7"
)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
# Fetch the top 10 global songs from a playlist
playlist_id = "6UeSakyzhiEt4NB3UAd6NQ"
results = sp.playlist_tracks(playlist_id)
```

```python
# Initialize empty lists for albums, artists, and songs
albums_data = []
artists_data = []
songs_data = []
```

```python
# Extract details
for track in results['items']:
    # Extract Track Info
    song = track['track']
    song_id = song['id']
    song_name = song['name']
    duration_ms = song['duration_ms']
    popularity = song['popularity']
    song_url = song['external_urls']['spotify']
    song_added = track['added_at']  # Timestamp when song was added to playlist

    # Extract Album Info
    album = song['album']
    album_id = album['id']
    album_name = album['name']
    album_release_date = album['release_date']
    album_total_tracks = album['total_tracks']
    album_url = album['external_urls']['spotify']

    # Extract Artist Info (Each song can have multiple artists)
    for artist in song['artists']:
        artist_id = artist['id']
        artist_name = artist['name']
        artist_url = artist['external_urls']['spotify']

        # Append artist data
        artists_data.append({
            "Artist_id": artist_id,
            "Artist_name": artist_name,
            "External_url": artist_url
        })

    # Append album data
    albums_data.append({
        "Album_id": album_id,
        "Name": album_name,
        "Release_date": album_release_date,
        "Total_tracks": album_total_tracks,
        "URL": album_url
    })

    # Append song data
    songs_data.append({
        "Song_id": song_id,
        "Song_name": song_name,
        "Duration_ms": duration_ms,
        "URL": song_url,
        "Popularity": popularity,
        "Song_added": song_added,
        "Album_id": album_id  # Foreign key linking to Album DataFrame
    })
```

```python
# Convert lists to Pandas DataFrames
df_albums = pd.DataFrame(albums_data).drop_duplicates(subset=['Album_id'])
df_artists = pd.DataFrame(artists_data).drop_duplicates(subset=['Artist_id'])
df_songs = pd.DataFrame(songs_data).drop_duplicates(subset=['Song_id'])
```
